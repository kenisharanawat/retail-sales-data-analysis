```python
import pandas as pd
import matplotlib.pyplot as plt

# sample dataset
data = pd.DataFrame({
    "Product": ["A", "B", "C", "D"],
    "Revenue": [5000, 7000, 4000, 6500]
})

# average revenue
avg_revenue = data["Revenue"].mean()
print("Average Revenue:", avg_revenue)

# top product
top_product = data.sort_values(by="Revenue", ascending=False)
print(top_product)

# visualization
plt.bar(data["Product"], data["Revenue"])
plt.title("Product Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```
